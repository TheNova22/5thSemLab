```python
import matplotlib.pyplot as plt
```

## Linear

```python
x = [x for x in range(-10,11)]
y = x.copy()
plt.plot(x,y,'r-')
plt.title('Linear')
plt.show()
```

## Binary Step

```python
def binaryStep(x):
    if x < 0 : return 0
    else: return 1
x , y = [val for val in range(-10,11)], [binaryStep(val) for val in x]
plt.plot(x,y,'r-')
plt.title('Binary Step')
plt.show()
```

```python
import math
def sigmoid(x,sigma):
    return 1/(1 + (math.e**(-sigma*x)))
x , y = [val for val in range(-10,11)], [sigmoid(val,1) for val in x]
plt.plot(x,y,'r-')
plt.title('Sigmoid (Sigma = 1)')
plt.show()
```

```python
def tanh(x,sigma):
    return (2 * sigmoid(x,1)) - 1
x , y = [val for val in range(-10,11)], [tanh(val,1) for val in x]
plt.plot(x,y,'r-')
plt.title('Bipolar Sigmoid (Sigma = 1)')
plt.show()
```
